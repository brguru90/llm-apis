```python
base64_txt=""

import base64

with open("stock3.png", "rb") as image_file:
    encoded_bytes = base64.b64encode(image_file.read())
    base64_txt = encoded_bytes.decode("utf-8")  


with open("base64.txt", "w") as text_file:
    text_file.write(base64_txt)
```

```python
import ollama

num_gpu=-1
kepp_llm_alive=60*5
num_ctx=1024*20 

response = ollama.chat(
    # model='llama3.2-vision',
    model='gemma3:12b',
    options={'num_ctx': num_ctx,'num_gpu': num_gpu},
    messages=[{
        'role': 'user',
        'content': 'Its a stock trend, Give me candle pairs which might trigger 100 point movement with maximum pullback of 40 points',
        'images': [base64_txt]
    }]
)

print(response.model_dump()["message"]["content"])
```

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import SystemMessage, HumanMessage
import imghdr
import base64

llm = ChatOllama(model='gemma3',temperature=0,num_ctx=num_ctx,num_gpu=num_gpu)

PROMPT = ChatPromptTemplate.from_messages(
    messages=[
        SystemMessage(content="You are an expert stock analyst. please respond to user queries"),
        HumanMessage(content=[
            {
                "type": "image_url",
                "image_url": f"data:image/png;base64,{base64_txt}"
            },
            {"type": "text", "text": "Give me candle pairs which might trigger 100 point movement with maximum pullback of 40 points"}
        ])
    ]
)
runnable = PROMPT | llm



# def file_to_data_uri_v2(path: str) -> str:
#     with open(path, 'rb') as f:
#         data = f.read()
#     fmt = imghdr.what(None, data)  # 'jpeg', 'png', etc.
#     ext = fmt or 'octet-stream'
#     mime_type = f'image/{ext}' if fmt else 'application/octet-stream'
#     encoded = base64.b64encode(data).decode('utf-8')
#     return f'data:{mime_type};base64,{encoded}'


# base64_txt = file_to_data_uri_v2("stock3.png")

new_chunk_summary = runnable.invoke({
    "img_base64": base64_txt,
}).content  

print("new_chunk_summary=", new_chunk_summary)

print(new_chunk_summary)
```
